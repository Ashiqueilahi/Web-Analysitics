```python
from github import Github
import time
import datetime

# Authenticate using your GitHub personal access token
g = Github("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")  # Replace with your actual token

# Get the NumPy repository
repo = g.get_repo("numpy/numpy")

# Get contributors
contributors = repo.get_contributors()
count = 0

# Print the top 100 contributors
for contributor in contributors:
    print(f"{contributor.login} ({contributor.contributions} contributions)")
    count += 1
    if count >= 100:
        break

# Handle rate limiting
if g.get_rate_limit().core.remaining < 10:
    reset_time = g.get_rate_limit().core.reset
    wait_time = (reset_time - datetime.datetime.now(datetime.timezone.utc)).total_seconds()
    print(f"Rate limit approaching. Sleeping for {wait_time:.2f} seconds...")
    time.sleep(wait_time + 1)
```
